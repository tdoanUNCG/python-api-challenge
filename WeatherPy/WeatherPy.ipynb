```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress

# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/hw_cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
# print(weather_api_key)
# cities
```

```python
# Perform API Calls
# Perform a weather check on each city using a series of successive API calls.
# Include a print log of each city as it's
# being processed (with the city number and city name).
url = "http://api.openweathermap.org/data/2.5/weather?"
#city = "new norfolk"
# build query URL
#query_url = url + "appid=" + weather_api_key + "&q=" + city
# print(query_url)
cloudiness = []
country = []
date = []
humidity = []
latitude = []
longitude = []
max_temp = []
wind_speed = []
ow_cities = []
```

```python
#weather_response = requests.get(query_url)
#weather_json = weather_response.json()
# get temperature from JSON response
# temperature = weather_json['main']['temp']
for city in cities:
    # build query URL
    query_url = url + 'appid=' + weather_api_key + '&q=' + city
    print(query_url)
    response = requests.get(query_url).json()
    if response['cod'] == '404':
        print(response['message'])
    else:
        #print(response['clouds']['all'])
        ow_cities.append(response['name'])
        cloudiness.append(response['clouds']['all'])
        country.append(response['sys']['country'])
        date.append(response['dt'])
        humidity.append(response['main']['humidity'])
        latitude.append(response['coord']['lat'])
        longitude.append(response['coord']['lon'])
        max_temp.append(response['main']['temp_max'])
        wind_speed.append(response['wind']['speed'])
    time.sleep(0.2)
    
#     response = requests.get(query_url)
#     weather_json = response.json()
#     weather_json['coord']['lat']
    #print(query_url)
#     # get weather data
     #weather_response = requests.get(query_url)
#     weather_json = weather_response.json()
#     # get temperature from JSON response
#     # temperature = weather_json['main']['temp']
#     # temperatures.append(temperature)
#     print(weather_json)
    
#print(weather_json)
#cloudiness
# convert k to f in max_temp list
for i in range(len(max_temp)):
    max_temp[i] = (max_temp[i]-273.15)*(9/5)+32

# convert wind speed to mph
for j in range(len(wind_speed)):
    wind_speed[j] = wind_speed[j]*2.237
```

```python
weather_dict = {
    "City": ow_cities,
    "Cloudiness": cloudiness,
    "Country": country,
    "Date": date,
    "Humidity": humidity,
    "Lat": latitude,
    "Lng": longitude,
    "Max Temp": max_temp,
    "Wind Speed": wind_speed
}
ow_weather_df = pd.DataFrame(weather_dict)
ow_weather_df
```

```python
# isolate northern and southern hemisphere cities
northern_hem = []
southern_hem = []
northern_hem_cities = []
southern_hem_cities = []
northern_hem_temp = []
southern_hem_temp = []
northern_hem_hum = []
southern_hem_hum = []
northern_hem_cloud = []
southern_hem_cloud = []
northern_hem_wind = []
southern_hem_wind = []
for m in range(len(longitude)):
    if longitude[m] > 0:
        northern_hem.append(longitude[m])
        northern_hem_cities.append(ow_cities[m])
        northern_hem_temp.append(max_temp[m])
        northern_hem_hum.append(humidity[m])
        northern_hem_cloud.append(cloudiness[m])
        northern_hem_wind.append(wind_speed[m])
    else:
        southern_hem.append(longitude[m])
        southern_hem_cities.append(ow_cities[m])
        southern_hem_temp.append(max_temp[m])
        southern_hem_hum.append(humidity[m])
        southern_hem_cloud.append(cloudiness[m])
        southern_hem_wind.append(wind_speed[m])
```

```python
#northern_hem
northern = {
    "City": northern_hem_cities,
    "Lat": northern_hem,
    "Max Temp": northern_hem_temp,
    "Humidity": northern_hem_hum,
    "Cloudiness": northern_hem_cloud,
    "Wind Speed": northern_hem_wind
}
north_df = pd.DataFrame(northern)

# southern hem
southern = {
    "City": southern_hem_cities,
    "Lat": southern_hem,
    "Max Temp": southern_hem_temp,
    "Humidity": southern_hem_hum,
    "Cloudiness": southern_hem_cloud,
    "Wind Speed": southern_hem_wind
}
south_df = pd.DataFrame(southern)

# plot northern hem lat vs temp
# add linear regression equation and line to plot
fig, ax = plt.subplots()
(slope, intercept, rvalue, pvalue, stderr) = linregress(north_df['Lat'],north_df['Max Temp'])
regress_values = north_df['Lat'] * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
ax.scatter(north_df['Lat'],north_df['Max Temp'],alpha=0.25)
ax.plot(north_df['Lat'],regress_values,"r-")
ax.annotate(line_eq,(50,15),fontsize=15,color='red')
ax.set_xlabel('Latitude')
ax.set_ylabel('Max Temp')
ax.set_title('Latitude vs. Temperature')
print(f"The r-squared value is: {rvalue}")
```
